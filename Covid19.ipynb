##Explpratory Data Analysis

Use the seaborn to explore the data.

Create a histogram of the state

```
sns.set_style('whitegrid')
co_data['state'].hist(bins=30)
plt.xlabel('state')
```

Let's create a jointplot to show 'state' vs 'cases'.




```
sns.jointplot(x='state',y='cases',data=co_data)
```

Let's create a jointplot showing the kde distributions of state vs deaths.

```
sns.jointplot(x='state',y='deaths',data=co_data)
```

Create a histogram of the date

```
sns.set_style('whitegrid')
co_data['date'].hist(bins=30)
plt.xlabel('date')
```
